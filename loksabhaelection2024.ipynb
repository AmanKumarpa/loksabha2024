# Total vote percentage in Parliamentry constituencies Name

```python
votes_sum = ad.groupby("ParliamentryName")[["Votes Secured Evm"]].sum()
nota = ad.groupby("ParliamentryName")[["Nota Votes Evm In AC"]].first()

total_df = votes_sum.merge(nota, on="ParliamentryName")

total_df["Total_Votes"] = total_df["Votes Secured Evm"] + total_df["Nota Votes Evm In AC"]

ad = ad.merge(total_df[["Total_Votes"]], left_on="ParliamentryName", right_index=True, how="left")
ad
```

```python
ad.rename(columns = {"Total_Votes_y":"Total_votes"}, inplace =True)
ad
```

**Now we find turnout percentage in general constituencies:**

**Turnout generalpercent in constituencies** = ( Total_votesTotal Electors in PC)*100

**Total Votes** = the sum of all votes cast in a Parliamentary Constituency (including all candidates and NOTA).

**Total Electors in PC** = the total number of registered voters in that constituency.

This gives the overall voter turnout percentage for each constituency. It represents the proportion of eligible voters who actually participated in the election

```python
ad["Turnout generalpercent"] = (ad["Total_Votes"]/ad["Total Electors In PC"])*100
```

# Vote Percentage by each party or Party Performance Analysis

```python
ad["Votesharepercentage"] = (ad["Votes Secured Evm"]/ad["Total_Votes"])*100
ad
```

## The column Votesharepercentage is calculated as:

**Votesharepercentage** = (Votes Secured EVM / Total_Votes)*100

**Votes Secured EVM**= the number of votes a candidate or party received through Electronic Voting Machines.

**Total_Votes** = the total number of registered voters in that Parliamentary Constituency (PC).

This gives the percentage of the total electorate that actually voted for a candidate/party, not just out of the people who cast their votes, but out of the entire eligible voting population.

# Marginal Victory

***Marginal victor*** = max(Votes Secured Evm) - 2ndmax(Votes Secured Evm)

```python
ad[ad["ParliamentryName"] == "Nawada"]
```

```python
# ✅ Step 1: Get total votes per Party per PC
party_totals = (
    ad.groupby(["ParliamentryName", "Party"], as_index=False)["Votes Secured Evm"]
      .sum()
      .rename(columns={"Votes Secured Evm": "TotalVotes_Party"})
)

# ✅ Step 2: Sort within each PC
party_totals = party_totals.sort_values(
    ["ParliamentryName", "TotalVotes_Party"], ascending=[True, False]
)

# ✅ Step 3: Take top 2 per PC
top2 = party_totals.groupby("ParliamentryName").head(2)

# ✅ Step 4: Compute margins (winner, runner-up, difference)
def compute_margin(g):
    g = g.reset_index(drop=True)  # ensure rows start at 0,1
    top1 = g.loc[0, "TotalVotes_Party"]
    top2_votes = g.loc[1, "TotalVotes_Party"] if len(g) > 1 else 0
    return pd.Series({
        "ParliamentryName": g.loc[0, "ParliamentryName"],
        "Winner_Party": g.loc[0, "Party"],
        "Winner_Votes": top1,
        "Runnerup_Party": g.loc[1, "Party"] if len(g) > 1 else None,
        "Runnerup_Votes": top2_votes,
        "Winning_Margin": top1 - top2_votes
    })

pc_margins = top2.groupby("ParliamentryName").apply(compute_margin).reset_index(drop=True)

# ✅ Step 5: Drop duplicate columns if already present
for col in ["Winner_Party", "Winner_Votes", "Runnerup_Party", "Runnerup_Votes", "Winning_Margin"]:
    if col in ad.columns:
        ad = ad.drop(columns=col)

# ✅ Step 6: Merge back with main dataset
ad = ad.merge(pc_margins, on="ParliamentryName", how="left")

# ✅ Final dataset
ad
```
