```python
import pandas as pd
import os
import sqlite3
from sqlalchemy import create_engine
import mysql.connector
```

```python
create = create_engine("sqlite:/// election.db")
```

```python
def ingest_db(file_path, table_name, conn):
    df1 = pd.read_csv(file_path)
    df1.to_sql(table_name, con=conn, if_exists="replace", index=False)
    print("Ingested:", table_name, df1.shape)   # moved inside function & fixed df name

conn = sqlite3.connect("election.db")

file_path = "C:/Users/amanp/Downloads/election commision.csv"
table_name = os.path.splitext(os.path.basename(file_path))[0]  

ingest_db(file_path, table_name, conn)
```

```python
tables= pd.read_sql_query("select name from sqlite_master where type ='table'",conn)
print(tables)
```

```python
for table in tables["name"]:
    print(pd.read_sql_query(f'SELECT count(*) AS rows FROM "{table}"', conn))
    print(pd.read_sql_query(f'SELECT * FROM "{table}" LIMIT 5', conn))  
```
